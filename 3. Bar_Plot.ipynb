```python
# Import necessary packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read data using pandas.
data=pd.read_csv('CompanySale.csv')
data.head()
```

# Bar plot (1)

```python
# Define the height and X labels for the bar plot.
height=np.array(data['shampoo']) # Extract the shampoo column as height.
x_labels=['Month '+str(i+1) for i in range(data.shape[0])] # Define the x labels using a list.

# Adjust the font family.
plt.rcParams['font.family']='Times New Roman'

# Create a figure.
plt.figure(figsize=(5,3),dpi=300)

# Create a simple bar plot.
plt.bar(x_labels,height,width=0.8,color='red')

# Add x and y ticks
plt.xticks(rotation=45)
plt.yticks([0,1000,2000,3000,4000],rotation=90)

# Add title
plt.title('Bar plot',fontsize=12,fontweight='bold')

# Add X and Y labels.
plt.xlabel('Month',fontsize=12,fontweight='bold')
plt.ylabel('Profit',fontsize=12,fontweight='bold')

# Display the plot.
plt.show()
```

# Bar plot (2)

```python
# Prepare the data.
sorted_indices= np.argsort(data['facecream']) # Get the indices that would sort the facecreame column.
sorted_facecream=np.array(data['facecream'])[sorted_indices] # Sort the facecream profit based on the sorted indices.
sorted_month=np.array(data['month_number'])[sorted_indices] # sort the corresponding month numbers based on the sorted indices.
sorted_month_new_facecream= ['Month '+str(m) for m in sorted_month] 

# Create a figure. 

# Create a figure.
plt.figure(figsize=(5,3),dpi=300)

# Create a simple bar plot.
plt.bar(sorted_month_new_facecream,sorted_facecream,width=0.6,color='navy')

# Add x and y ticks
plt.xticks(rotation=45)
plt.yticks([0,1000,2000,3000,4000],rotation=90)

# Add title
plt.title('Bar plot',fontsize=12,fontweight='bold')

# Add X and Y labels.
plt.xlabel('Month',fontsize=12,fontweight='bold')
plt.ylabel('Profit',fontsize=12,fontweight='bold')

# Add legend.
plt.legend(['Facecream profit'])

# Display the plot.
plt.show()
```

# Bar plot (3)

```python
# Define data for horizontal bar plot.
X=np.array(data['shampoo']) # Extract the shampoo column as X.
y_labels=['Month '+str(i+1) for i in range(data.shape[0])] # Define the y labels using a list.

# Create a figure.
plt.figure(figsize=(5,3),dpi=300)

# Create a simple bar plot.
plt.barh(y_labels,X, color='blue')

# Add title
plt.title('Horizontal bar plot',fontsize=12,fontweight='bold')

# Add X and Y labels.
plt.ylabel('Month',fontsize=12,fontweight='bold')
plt.xlabel('Profit',fontsize=12,fontweight='bold')

# Add legend.
plt.legend(['shampoo profit'])

# Display the plot.
plt.show()
```

```python
# Group the data by month and sum the profit for all products.
monthly_profit = data.groupby('month_number').sum()

# Get the months and corresponding profits.
months = monthly_profit.index
facrcream = monthly_profit['facecream']
facewash = monthly_profit['facewash']
toothpaste = monthly_profit['toothpaste']
bathingsoap = monthly_profit['bathingsoap']
shampoo = monthly_profit['shampoo']
moisturizer = monthly_profit['moisturizer']

x_labels=['Month '+str(i+1) for i in range(data.shape[0])] # Define the x labels using a list.


# Define width of each bar.
bar_width=0.35

# Define the position for each bar.
bar_position = np.arange(len(months))

# Create a figure with custom dpi and size. 
plt.figure(figsize=(9,6),dpi=300)

# Create a bar plot. 
plt.bar(bar_position,facrcream,width=bar_width,color='teal',label='facrcream')
plt.bar(bar_position,facewash,width=bar_width,color='orange',label='facewash',
       bottom=facrcream)
plt.bar(bar_position,toothpaste,width=bar_width,color='navy',label='toothpaste',
       bottom=facewash+facrcream)
plt.bar(bar_position,bathingsoap,width=bar_width,color='green',label='bathingsoap',
       bottom=facewash+facrcream+toothpaste)
plt.bar(bar_position,shampoo,width=bar_width,color='red',label='shampoo',
       bottom=facewash+facrcream+toothpaste+bathingsoap)
plt.bar(bar_position,moisturizer,width=bar_width,color='purple',label='moisturizer',
       bottom=facewash+facrcream+toothpaste+bathingsoap+shampoo)


# Add legend
plt.legend()

# Add title
plt.title('Bar plot',fontsize=12,fontweight='bold')

# Add X and Y labels.
plt.xlabel('Month',fontsize=12,fontweight='bold')
plt.ylabel('Total Profit',fontsize=12,fontweight='bold')

# Add ticks.
plt.xticks([i for i in range(12)],x_labels,rotation=45)
plt.yticks(rotation=90)
# Display the plot.
plt.show()
```

# Bar plot (5)

```python
# Group the data by month and sum the profit for all products.
monthly_profit = data.groupby('month_number').sum()

# Get the months and corresponding profits.
months = monthly_profit.index
facrcream = monthly_profit['facecream']
facewash = monthly_profit['facewash']
toothpaste = monthly_profit['toothpaste']
bathingsoap = monthly_profit['bathingsoap']
shampoo = monthly_profit['shampoo']
moisturizer = monthly_profit['moisturizer']

x_labels=['Month '+str(i+1) for i in range(data.shape[0])] # Define the x labels using a list.


# Define the bar width.
bar_width=0.1
# Define the position for each bar.
bar_position = np.arange(len(months))

# Create a figure with custom dpi and size. 
plt.figure(figsize=(9,6),dpi=300)

# Create a bar plot. 
plt.bar(bar_position-3*bar_width,facrcream,width=bar_width,color='teal',label='facrcream')
plt.bar(bar_position-2*bar_width,facewash,width=bar_width,color='orange',label='facewash')
plt.bar(bar_position-bar_width,toothpaste,width=bar_width,color='navy',label='toothpaste')
plt.bar(bar_position,bathingsoap,width=bar_width,color='red',label='bathingsoap')
plt.bar(bar_position+bar_width,shampoo,width=bar_width,color='green',label='shampoo')
plt.bar(bar_position+2*bar_width,moisturizer,width=bar_width,color='magenta',label='moisturizer')

# Add legend
plt.legend()


# Add title
plt.title('Bar plot',fontsize=12,fontweight='bold')

# Add X and Y labels.
plt.xlabel('Month',fontsize=12,fontweight='bold')
plt.ylabel('Total Profit',fontsize=12,fontweight='bold')

# Add ticks.
plt.xticks([i for i in range(12)],x_labels,rotation=45)
plt.yticks(rotation=90)

# Display the plot
plt.show()
```
