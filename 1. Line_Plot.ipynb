```python
# Import necessary packages
import matplotlib.pyplot as plt
import numpy as np
```

# Line plot example (1)

```python
# Create a sample data.
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot using matplotlib
plt.plot(x,y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot example')

# Display the plot.
plt.show()
```

# Line plot example (2)

```python
# Create sample data using numpy library.
x = np.linspace(0, 2*np.pi,100) # Returns 100 evenly spaced values between 0 and 2*pi
y = np.sin(x) # Compute the Sine function for each value of x.
 
# Adjust the font in matplotlib library
plt.rcParams['font.family']='Times New Roman'

# Create a figure with specific customization.
plt.figure(figsize=(3,2),dpi=300)

# Create a line plot
# Line style list: ['solid','dotted','dashed','dashdot']
plt.plot(x,y, color='maroon',linestyle='dashdot',linewidth=0.3, marker='.',markersize=2)

# Add x and y labels
plt.xlabel('X-axis',fontsize=10,fontweight='bold')
plt.ylabel('Y-axis',fontsize=10,fontweight='bold')

# Add title
plt.title('Sine Curve',fontsize=10,fontweight='bold')

# Add legend
plt.legend(['Sine Function'])

# Add grid
plt.grid(True,linewidth=0.2)

# Limit the x and y axes.
plt.xlim(0,6)
plt.ylim(-1,1)

# Modify the x and y ticks
plt.xticks([0,1,2,3,4,5,6])
plt.yticks([-1,-0.5,0,0.5,1])

# Display the plot.
plt.show()
```

# Line plot example (3)

```python
# Create sample data using numpy lobrary.
x = np.linspace(0, 2*np.pi,100) # Returns 100 evenly spaced values between 0 and 2*pi
y_sine = np.sin(x) # Compute the Sine function for each value of x.
y_cosine = np.cos(x) # Compute the Cosine function for each value of x.
y_product = y_sine*y_cosine # Compute the product od sine and cosine functions.

# Adjust the font in matplotlib library
plt.rcParams['font.family']='Times New Roman'

# Create a figure with specific customization.
plt.figure(figsize=(5,3),dpi=300)

# Create a line plot
# Line style list: ['solid','dotted','dashed','dashdot']
plt.plot(x,y_sine, color='green',linestyle='solid',linewidth=2,label='Sine')
plt.plot(x,y_cosine, color='blue',linestyle='solid',linewidth=2,label='Cosine')
plt.plot(x,y_product, color='orange',linestyle='solid',linewidth=2,label='Sine*Cosine')

# Add legend
plt.legend()

# Add grid
plt.grid(True,linewidth=0.2)

# Limit the x and y axes.
plt.xlim(0,2*np.pi)
plt.ylim(-1,1)

# Modify the x and y ticks
plt.xticks([0,np.pi/2,np.pi,1.5*np.pi,2*np.pi],['0','pi/2','pi','3pi/2','2pi'])
plt.yticks([-1,-0.5,0,0.5,1])

# Add x and y labels
plt.xlabel('X-axis',fontsize=10,fontweight='bold')
plt.ylabel('Y-axis',fontsize=10,fontweight='bold')

# Add title
plt.title('Multiple plots',fontsize=10,fontweight='bold')

# Display the plot
plt.show()
```
