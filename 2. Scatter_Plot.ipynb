```python
# Import packages
import matplotlib.pyplot as plt
import numpy as np
```

# Scatter plot (1)

```python
# Create sample data
x = np.linspace(0, 2*np.pi,50) # Return 50 evenly spaced numbers between 0 and 2p.
y = np.sin(x) # Compute the sine function for each value of x.

# Adjust the font
plt.rcParams['font.family']='Times New Roman'

# Create a figure.
plt.figure(figsize=(4,2),dpi=300)

# Create scatter plot.
plt.scatter(x,y, color='green',marker='.',s=1,label='Sine curve')

# Add labels and title.
plt.xlabel('X-axis',fontsize=12,fontweight='bold')
plt.ylabel('Y-axis',fontsize=12,fontweight='bold')
plt.title('Scatter plot',fontsize=12,fontweight='bold')

# Add legend
plt.legend()

# Add grid.
plt.grid(True,linewidth=0.2)

# Limit the x and y axes.
plt.xlim(0,2*np.pi)
plt.ylim(-1,1)

# Add X ticks
plt.xticks([0, np.pi/2,np.pi,1.5*np.pi,2*np.pi],['0','P/2','P','3P/2','2P'])

# Display the plot.
plt.show()
```

# Scatter plot (2)

```python
# Create sample data.
x = np.random.rand(100) # Generate 100 random values between 0 and 1 for the x-axis.
y = np.random.rand(100) # Generate 100 random values between 0 and 1 for the y-axis.

sizes = np.random.uniform(1,50,100) # Generate 100 random values between 1 and 50.
colors = np.random.uniform(1,50,100) # Generate 100 random colors between 1 and 50.

# Create a figure.
plt.figure(figsize=(4,2),dpi=300)

# Create scatter plot.
plt.scatter(x,y,c=colors,marker='.',s=sizes,cmap='jet',label='Points')

# Add labels and title.
plt.xlabel('X-axis',fontsize=12,fontweight='bold')
plt.ylabel('Y-axis',fontsize=12,fontweight='bold')
plt.title('Scatter plot',fontsize=12,fontweight='bold')

# Add grid.
plt.grid(True,linewidth=0.2)

# Limit the x and y axes.
plt.xlim(0,1)
plt.ylim(0,1)

# Add legend
plt.legend()

# Dispaly the plot.
plt.show()
```
