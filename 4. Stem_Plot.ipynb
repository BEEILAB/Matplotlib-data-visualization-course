```python
# Import necessary packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create sample data.
x = np.linspace(0,2*np.pi,40) # Generate 20 evenly spaced values between 0 and 2P.
y = np.sin(x) # Compute the sine of each x value.
```

```python
# font adjustment.
plt.rcParams['font.family']='Times New Roman'

# Create a figure with custom dpi and size.
plt.figure(figsize=(5,3),dpi=300)

#Create a stem plot.
# linefmt list=['-',':','--','-.']
# markerfmt list=['o','D','*','s','x','+','p','d','>','^','h','H','v']
plt.stem(x,y,linefmt='-',markerfmt='H')

# Add labels and title.
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stem plot')

# Limit the x axis and y axis.
plt.xlim(0,2*np.pi)
plt.ylim(-1,1)

# Add ticks
plt.yticks([-1,-0.5,0,0.5,1])
plt.xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi],['0','P/2','P','3P/2','2P'])

# Add grid.
plt.grid(True,linewidth=0.1)

# Add legend.
plt.legend(['Sine'])

# Display the plot.
plt.show()
```
